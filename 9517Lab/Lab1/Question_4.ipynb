```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('cat.png', 0)
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
sigma = 1.0
image_L = cv2.GaussianBlur(img, (11, 11), sigma)
# plot the image which convolues with a Gaussian kernel
plt.imshow(image_L, cmap='gray', vmin=0, vmax=255)
```

```python
image_H = cv2.subtract(img, image_L)
# the blurred image L is subtracted from the input image I, resulting in image H
plt.imshow(image_H, cmap='gray')
```

```python
alpha = 1.25
image_O = cv2.addWeighted(image_H, alpha, img, 1, 0)
plt.imshow(image_O, cmap='gray', vmin=0, vmax=255)
```

```python
# contrast stretching function in Question 1
def contrast_stretching(image):
    a = 0
    b = 255
    c = np.min(image)
    d = np.max(image)
    new_image = (image - c) * ((b - a) / (d - c)) + a
    return new_image
```

```python
# plot image I after contrast stretching
I = contrast_stretching(img)
plt.imshow(I, cmap='gray', vmin=0, vmax=255)
```

```python
# plot image O after contrast stretching
O = contrast_stretching(image_O)
plt.imshow(O, cmap='gray', vmin=0, vmax=255)
```
